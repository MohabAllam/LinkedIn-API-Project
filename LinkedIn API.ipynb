```python
import csv

with open('LinkedIn.csv', 'w', newline= '') as file:

    writer = csv.DictWriter(file, fieldnames= ['Total Reactions', 'Name', 'Company Name'])
    writer.writeheader()

    for d in data['data']:
        total_reactions = d['totalReactionCount']
        name = d['author'].get('username')
        company_name = d['company'].get('name')

        writer.writerow({'Total Reactions' : total_reactions, 'Name' : name, 'Company Name' : company_name})
```
